```python
import altair as alt
alt.data_transformers.enable('default', max_rows=None)

df_doc_viz = pd.read_csv("topic_model_viz/doc_vectors_viz.csv", index_col="doc_id")

df_doc_viz.rename(columns={'Unnamed: 0': 'doc_id'}, inplace=True)
df_doc_viz.columns

# Altair scatter plot
scatter_plot = alt.Chart(df_doc_viz
).mark_circle().encode(
    x='doc_x',
    y='doc_y',
    tooltip=['doc_id'],
    color='genre:N'
).properties(
    width=1200,
    height=800
).interactive()

scatter_plot.show() 
```
