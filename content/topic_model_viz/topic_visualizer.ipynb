```python
# import altair with an abbreviated alias
# pip install "vegafusion[embed]>=1.5.0"
import altair as alt
# alt.data_transformers.enable("vegafusion")

df_doc_viz = pd.read_csv("doc_vectors_viz.csv", index_col="doc_id")
# Altair does not allow > 5K points, otherwise is needs pyarrow which is not supported here
df_doc_viz = df_doc_viz.sample(n=5000)
df_doc_viz.rename(columns={'Unnamed: 0': 'doc_id'}, inplace=True)
df_doc_viz.columns

# Altair scatter plot
scatter_plot = alt.Chart(df_doc_viz
).mark_circle().encode(
    x='doc_x',
    y='doc_y',
    tooltip=['doc_id'],
    color='genre:N'
).properties(
    width=1200,
    height=800
).interactive()

scatter_plot.show() 
```
